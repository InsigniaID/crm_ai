````python
def get_combinations(number_of_pairs = 2):
    combinations = list(itertools.combinations(contents, number_of_pairs))
    result = []
    for item1 in combinations:
        for item2 in metrics:
            result.append((list(item1), item2))
    return result
````

````python
df_results: list[pd.DataFrame] = []
for i in get_combinations(number_of_pairs=2):
    list_of_columns = i[0]
    columns = ','.join(list_of_columns)
    value = i[1]
    df_temp = sqldf.run(f"""select {columns}, {value} from df where {value} is not null group by {columns}""")
    df_max = sqldf.run(f"""select {columns}, {value} as max_{value} from df_temp order by {value} desc limit 1""")
    df_min = sqldf.run(f"""select {columns}, {value} as min_{value} from df_temp order by {value} asc limit 1""")
    # print(df_temp.to_string())
    df_results.extend([df_temp.to_string(), df_max.to_string(), df_min.to_string()])
````

````python
def get_completion(prompt, model="gpt-3.5-turbo-16k"): # Andrew mentioned that the prompt/ completion paradigm is preferable for this class
    messages = [{"role": "user", 
                 "content": prompt
                }]
    response = openai.ChatCompletion.create(
        model=model,
        messages=messages,
        temperature=0, # this is the degree of randomness of the model's output,
        max_tokens = 1000
    )
    return response.choices[0].message['content']
````
